```python
img=cv2.imread('light.jpg')
HSV=cv2.cvtColor(img,cv2.COLOR_RGB2HSV)
# 蓝色的范围，不同光照条件下不一样，可灵活调整
# 绿色的范围
lower_green = np.array([35, 20, 35])
upper_green = np.array([70, 255, 255])
# 红色的范围
low_red = np.array([50, 80, 90])
high_red = np.array([255, 255, 255])

# 3.inRange()：介于lower/upper之间的为白色，其余黑色
 #将低于lower_blue和高于upper_blue的部分变成0，中间部分变为255（本文中即蓝色部分为255）
# Step3. 形态学运算，膨胀

mask= cv2.inRange(HSV, low_red, high_red)
dilated= cv2.dilate(mask, cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (3, 3)), iterations=2)

# 4.只保留原图中的蓝色部分
 #mask ≠0时frame与frame按位相与，即显示蓝色部分，其他部分为0（黑色）
res = cv2.bitwise_and(img, img, mask=dilated)
cv2.imshow('frame', img)
# cv2.imshow('mask', mask)
cv2.imshow('res', res)
cv2.waitKey(0)  
cv2.destroyAllWindows()
```

```python
# More: http://ex2tron.top

import cv2
import numpy as np

# HSV中
# Blue：[[[120 255 255]]]
# Green：[[[ 60 255 255]]]
# Red：[[[  0 255 255]]]

capture = cv2.VideoCapture(1)

# 蓝色的范围
lower_blue = np.array([100, 110, 110])
upper_blue = np.array([130, 255, 255])

# 绿色的范围
lower_green = np.array([40, 90, 90])
upper_green = np.array([70, 255, 255])

# 红色的范围
lower_red = np.array([160, 120, 120])
upper_red = np.array([179, 255, 255])

while(True):
    # 1.捕获视频中的一帧
    ret, frame = capture.read()

    # 2.从BGR转换到HSV
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

    mask_blue = cv2.inRange(hsv, lower_blue, upper_blue)
    mask_green = cv2.inRange(hsv, lower_green, upper_green)
    mask_red = cv2.inRange(hsv, lower_red, upper_red)

    # 3.将所有的mask相加，就可以同时显示了
    mask = mask_blue + mask_green + mask_red

    # 4.保留原图中的三种颜色部分
    res = cv2.bitwise_and(frame, frame, mask=mask)

    cv2.imshow('frame', frame)
    cv2.imshow('mask', mask)
    cv2.imshow('res', res)

    if cv2.waitKey(1) == ord('q'):
        break
```

```python
img=cv2.imread('light2.jpg')
HSV=cv2.cvtColor(img,cv2.COLOR_RGB2HSV)
# 蓝色的范围，不同光照条件下不一样，可灵活调整
# 绿色的范围
lower_green = np.array([35, 20, 147])
upper_green = np.array([70, 255, 255])

# 红色的范围
low_red = np.array([0, 56, 90])
high_red = np.array([9, 203, 255])

# 3.inRange()：介于lower/upper之间的为白色，其余黑色
 #将低于lower_blue和高于upper_blue的部分变成0，中间部分变为255（本文中即蓝色部分为255）
mask = cv2.inRange(HSV, low_red, high_red)

# 4.只保留原图中的蓝色部分
 #mask ≠0时frame与frame按位相与，即显示蓝色部分，其他部分为0（黑色）
res = cv2.bitwise_and(img, img, mask=mask)
cv2.imshow('frame', img)
cv2.imshow('mask', mask)
cv2.imshow('res', res)
cv2.waitKey(0)  
cv2.destroyAllWindows()

###################################################

#载入并显示图片
# img=cv2.imread('light.jpg')
# cv2.imshow('img',img)
#降噪（模糊处理用来减少瑕疵点）
blur = cv2.blur(res, (5,5))
#灰度化
gray=cv2.cvtColor(blur,cv2.COLOR_BGR2GRAY)
#输出图像大小，方便根据图像大小调节minRadius和maxRadius
#霍夫变换圆检测
canny = cv2.Canny(gray, 40, 80,) 
cv2.imshow('canny',canny)
circles= cv2.HoughCircles(canny,cv2.HOUGH_GRADIENT,1,100,param1=100,param2=70,minRadius=0,maxRadius=130)
# #输出返回值，方便查看类型

# 先排个序,确定最上面的是红灯，中间的是黄灯，最后的是绿灯
print(type(circles))
# print(circles.shape)
if circles !=None:
    circles=circles[0].copy()

# if len(circles) !=0:
  
    idex=np.lexsort([circles[:,1]])#
    circles=circles[idex,:]
    print('排序以后:',circles)
    print('圆的个数：',len(circles))

    #根据检测到圆的信息，画出每一个圆
    rectangles=[]
    for circle in circles:
        rectangle=[]
        #圆的基本信息
        print('圆心'+str(circle),circle[2])
        #坐标行列
        x=int(circle[0])
        y=int(circle[1])
        #半径
        r=int(circle[2])
        l=int(circle[0]-circle[2])
        t=int(circle[1]-circle[2])
    # 画方框
        cv2.rectangle(img, (l, t), (l + 2*r, t +2*r), (0, 255, 0), 2)
    #记得截取图片时，是（y:y+h,x,x+w,:）    
        rectangles.append(img[t:t+2*r,l:l+2*r,:])
        img=cv2.circle(img,(x,y),r,(0,0,255),-1)
    #显示新图像
    cv2.imshow('res',img)


    '''
    HSV改变亮度和饱和度
    色调H：用角度度量，取值范围为0°～360°，
    从红色开始按逆时针方向计算，红色为0°，绿色为120°,蓝色为240°。它们的补色是：黄色为60°，青色为180°,品红为300°；
    饱和度S：取值范围为0.0～1.0；
    亮度V：取值范围为0.0(黑色)～1.0(白色)。
    '''

    # 把亮度排序,所传入的rectangle是按红黄绿排好序的
    lightness=[]
    for i in range(len(rectangles)):
    #打印每个截取图片的大小
        print(rectangles[i].shape)
        HSV = cv2.cvtColor(rectangles[i], cv2.COLOR_BGR2HSV)
        H, S, V = cv2.split(HSV)
        V=np.mean(V)
        lightness.append(('color'+str(i),V))
    #     print(V,'V')
    #     plt.imshow(rectangles[i])
    #     plt.show()

    print(lightness,'lightness_1')
    lightness.sort(key=lambda x:x[1])
#     print(lightness,'lightness_2')
    #按任意键退出
cv2.waitKey(0)
cv2.destroyAllWindows()
```
